```python
import numpy as np
```

```python
k = 200    #num of feature vectors
d = 10     #dim of feature vectors
s = 4       #size of prototype group

R = np.random.binomial(p = 0.2, n = 1, size = (d,1))
R = R / np.linalg.norm(R)
C = R.dot(R.transpose()) 
C = np.ones(d)
feat = np.random.multivariate_normal(np.zeros(d), C, size = k)
idx, sum_ = zip(*sorted(zip(range(d), feat.sum(axis = 0)), key = lambda x : x[1], reverse = True))
```

```python
C
```

```python
prototype_idx = np.array(idx[0:s])
prototype_feat = feat[:, prototype_idx]
```

```python
p = 1.0 - 0.01
S = feat.transpose().dot(feat)
Q = (p + (1-p) * np.diag(S.diagonal())) * S
R = feat.transpose().dot(prototype_feat)
W = R.transpose().dot(np.linalg.inv(Q))
```

```python
z = W.dot(feat.transpose()).transpose()
```
